```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("../data/weather_data.db")

df = pd.read_sql_query(
    "SELECT * FROM weather_data ORDER BY extraction_time ASC",
    conn
)

conn.close()

df
```

```python
df["extraction_time"] = pd.to_datetime(df["extraction_time"])
df.dtypes
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["extraction_time"], df["temperature"], marker="o")
plt.title("Temperature Over Time")
plt.xlabel("Extraction Time")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["extraction_time"], df["humidity"], marker="o")
plt.title("Humidity Over Time")
plt.xlabel("Extraction Time")
plt.ylabel("Humidity (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["extraction_time"], df["wind_speed"], marker="o")
plt.title("Wind Speed Over Time")
plt.xlabel("Extraction Time")
plt.ylabel("Wind Speed")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
